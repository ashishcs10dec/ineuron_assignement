```python
#2.  Discuss the concept of single inheritance and multiple inheritance, highlighting their differences and advantages.
#Ans. Single inheritance means inherit only one class into another call but in multiple inheritance we can inherit more than
#one class into another one. Multiple inheritance will provide more code reusability and maintance of code will be easy.

#Single inheritance
class Person():
    def __init__(self,name):
        self.name=name
        
    def display(self):
        print(f'Hello {self.name}')
        
class Employee(Person):
    def __init__(self,name,age):
        self.name=name
        self.age=age
        super().__init__(name)
        
    def display(self):
        super().display()
        print(f"Hello {self.name}, your age is {self.age}")
        
emp=Employee('Ashish',30)
emp.display()

#Multiple inheritance
class Gender():
    def __init__(self,gender):
        self.gender=gender
        
    def display_gender(self):
        print(f"Gender is {self.gender}")

class Person():
    def __init__(self,name):
        self.name=name
        
    def display(self):
        print(f'Hello {self.name}')
        
class Employee(Gender,Person):
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
        super().__init__(name)
        super().__init__(gender)
        
    def display(self):
        super().display()
        super().display_gender()
        print(f"Hello {self.name} ({self.gender}), your age is {self.age}")
        
emp=Employee('Ashish',30,'Male')
emp.display()
```

```python
#3. Explain the terms "base class" and "derived class" in the context of inheritance.
#Ans. Base class is the parent class means which class will be inharit and derived class is the child class means which class is
#going to use base class. Let's understand via an example:

class Person():
    def __init__(self):
        pass
    
class Employee(Person):
    def __init__(self):
        pass

#In the above example, Person is the base class because it is going to be inharit in Employee class. So Employee class is our
#derived class.
```

```python
#5. What is the purpose of the "super" keyword in inheritance? Provide an example.
#Ans. Super keyword is used to call the methods or functions of parent class into child class.
#e.g.

class Person():
    def __init__(self):
        pass
    def display(self):
        pass
    
class Employee(Person):
    def __init__(self):
        super().__init__()
        pass
    def display(self):
        super().display()
        pass
    
#As we can see in the above example, Person class init and display methods are called in Employee class by using super keyword.
```

```python
#6. Create a base class called "Vehicle" with attributes like "make", "model", and "year".
#Then, create a derived class called "Car" that inherits from "Vehicle" and adds an
#attribute called "fuel_type". Implement appropriate methods in both classe
#Ans.

class Vehicle():
    def __init__(self,make, model, year):
        self.make=make
        self.model=model
        self.year=year
        print(f"{self.model} car is made by {self.make} in the year of {self.year}")
        
class Car(Vehicle):
    def __init__(self,make,model,year,fuel_type):
        super().__init__(make, model, year)
        self.fuel_type=fuel_type
        print(f"Fuel type of this car is {self.fuel_type}")
        
car=Car("Suzuki","Breeza","2023","Petrol")
```

```python
#7. Create a base class called "Employee" with attributes like "name" and "salary."
#Derive two classes, "Manager" and "Developer," from "Employee." Add an additional
#attribute called "department" for the "Manager" class and "programming_language"
#for the "Developer" class.
#Ans. 

class Employee():
    def __init__(self, name, salary):
        self.name=name
        self.salary=salary
        
    def display_info(self):
        print(f"Hello {self.name}! your salary is {self.salary}")
        
class Manager(Employee):
    def __init__(self, name, salary, department):
        self.name=name
        self.salary=salary
        self.department=department
        super().__init__(name, salary)
        
    def display_info(self):
        super().display_info()
        print(f"Hello {self.name}! your salary is {self.salary} under {self.department}")
        
class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        self.name=name
        self.salary=salary
        self.programming_language=programming_language
        super().__init__(name, salary, programming_language)
        
    def display_info(self):
        super().display_info()
        print(f"Hello {self.name}! your salary is {self.salary} and your programming language is {self.department}")
        
dev=Manager('Jack','400000','Physics')
dev.display_info()

dev=Manager('Clark','500000','C++')
dev.display_info()
```

```python
#8. Design a base class called "Shape" with attributes like "colour" and "border_width."
#Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
#specific attributes like "length" and "width" for the "Rectangle" class and "radius" for
#the "Circle" class.
#Ans.

class Shape():
    def __init__(self, color, border_width):
        self.color=color
        self.border_width=border_width
        
    def display_info(self):
        print(f"{self.color} color has {self.border_width} width")
        
class Rectangle(Shape):
    def __init__(self,color, border_width, length, width):
        self.color=color
        self.border_width=border_width
        self.length=length
        self.width=width
        super().__init__(color, border_width)
        
    def display_info(self):
        super().display_info()
        print(f"{self.color} color has {self.border_width} width with lenth & width {self.length} & {self.width}")
              
class Circle(Shape):
    def __init__(self,color, border_width, radius):
        self.color=color
        self.border_width=border_width
        self.radius=radius
        super().__init__(color, border_width)
              
    def display_info(self):
        super().display_info()
        print(f"{self.color} color has {self.border_width} width with radius {self.radius}")
              
rec=Rectangle("red","10px","100px","120px")
rec.display_info()
            
cir=Circle("red","10px","360")
cir.display_info()
        
```

```python
#9. Create a base class called "Device" with attributes like "brand" and "model." Derive
#two classes, "Phone" and "Tablet," from "Device." Add specific attributes like
#"screen_size" for the "Phone" class and "battery_capacity" for the "Tablet" class.
#Ans.

class Device():
    def __init__(self,brand, model):
        self.brand=brand
        self.model=model
        
    def display_info(self):
        print(f"Brand - {self.brand} , Model - {self.model}")
        
class Phone(Device):
    def __init__(self,brand, model, screen_size):
        self.brand=brand
        self.model=model
        self.screen_size=screen_size
        super().__init__(brand,model)
        
    def display_info(self):
        super().display_info()
        print(f"{self.model} is a product of {self.brand} with screen size {self.screen_size}")
        
class Tablet(Device):
    def __init__(self,brand, model, battery_capacity):
        self.brand=brand
        self.model=model
        self.battery_capacity=battery_capacity
        super().__init__(brand,model)
        
    def display_info(self):
        super().display_info()
        print(f"{self.model} is a product of {self.brand} with battery {self.battery_capacity}")        
        
phone=Phone("Apple","iPhon 14 Pro Max","6'")
phone.display_info()

tablet=Tablet("Apple","iPad","5000mAH")
tablet.display_info()
```

```python
#10. Create a base class called "BankAccount" with attributes like "account_number" and
#"balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
#"BankAccount." Add specific methods like "calculate_interest" for the
#"SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.
#Ans.

class BankAccount():
    def __init__(self,account_number,balance):
        self.account_number=account_number
        self.balance=balance
    def display_info(self):
        print(f"This account {self.account_number} has Rs. {self.balance}")
        
class SavingsAccount(BankAccount):
    def __init__(self,account_number,balance,interest_rate):
        self.account_number=account_number
        self.balance=balance
        self.interest_rate=interest_rate
        super().__init__(account_number,balance)        
    def calculate_interest(self):
        super().display_info()
        amount=float(self.balance)*float(self.interest_rate)/100
        print(f"The interest amount is {amount} at the rate of {self.interest_rate}")
        
class CheckingAccount(BankAccount):
    def __init__(self,account_number,balance,fees):
        self.account_number=account_number
        self.balance=balance
        self.fees=fees
        super().__init__(account_number,balance)        
    def deduct_fees(self):
        super().display_info()
        amount=float(self.balance) - float(self.fees)
        print(f"Remaining amount is {amount} because {self.fees} Rs. is deducted as bank charges")
        
sav=SavingsAccount("1009293222",100000,4)
sav.calculate_interest()

checking=CheckingAccount("1009293222",100000,400)
checking.deduct_fees()
        
```
